What are the unique countries present in the dataset?

How many times does each country appear in the dataset?

Create a Pie chart for top 3 countries ?

#### Observation : Zomato maximum records or transaction are from India.After that USA and then United Kingdom

How many ratings are there for each combination of aggregate rating, rating color, and rating text?

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

Observation :
1. When rating is between 4.5 to 4.9 --> Excellent
2. When rating is between 4.0 to 4.4 --> Very Good
3. When rating is between 3.5 to 3.9 --> Good
4. When rating is between 2.5 to 3.4 --> Average
5. When rating is between 1.8 to 2.4 --> Poor

See the first 5 records?

```python
ratings.head()
```

What is the distribution of restaurants based on their aggregate rating?

```python
### Bar Plot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,palette="viridis",ci=0)
plt.show();
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'],ci=0)
plt.show()
```

#### Observation : 
##### 1. Not Rated count is very high
##### 2. Maximum number of rating are between 2.5 to 3.4

How are restaurants distributed based on their rating color categories?

```python
### Count Plot
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
plt.show()
```
